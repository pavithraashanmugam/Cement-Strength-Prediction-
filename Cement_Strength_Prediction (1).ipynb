importing libraries

```python
# import required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

load datas from csv to dataframe

```python
#load data
data =  pd.read_csv('/content/concrete_data.csv')
```

```python
#print 1st five rows
data.head()
```

```python
# check if there is any null values
data.isnull().sum()
```

```python
data.describe()
```

```python
print(data.dtypes)
```

Split data into features & target

```python
# split dataset into features & target datas for training & testing
features = data.drop(columns='Strength', axis = 1)
target = data['Strength']
print(features)
print(target)
```

Standardize the data

```python
# create instance for standardizing the data
scaler = StandardScaler()
```

```python
# standardize the feature datas
std_features = scaler.fit_transform(features)
print(std_features)
```

Reassigning the std_features to features

```python
features = std_features
target = data['Strength']
print(features)
print(target)
```

Train-Test-Split

```python
# split datas for testing and training
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size = 0.2, random_state = 2)
print(features.shape,X_train.shape,X_test.shape)
```

```python
#create instance for linear regression
model = LinearRegression()
```

```python
# fitting the data to the model
model.fit(X_train,Y_train)
```

Model Evaluation

```python
# get the mean_squared_error value. (lesser the value, the performance of the model is good)
X_train_prediction = model.predict(X_train)
training_data_mse = mean_squared_error(Y_train, X_train_prediction)
print("Training data MSE is :", training_data_mse)

X_test_prediction = model.predict(X_test)
testing_data_mse = mean_squared_error( Y_test, X_test_prediction)
print("Testing data MSE is : ", testing_data_mse)
```

```python
# r2 score prediction. (value closer to 1, the performance is good ; value closer to 0, the performance id poor)
X_train_prediction = model.predict(X_train)
training_data_r2 = r2_score(Y_train, X_train_prediction)
print("Training data r2_score is :", training_data_r2)

X_test_prediction = model.predict(X_test)
testing_data_r2 = r2_score( Y_test, X_test_prediction)
print("Testing data r2_score is : ", testing_data_r2)
```

Since the linear Regression is not giving good prediction rate, we are using other model for prediction.
In this case XGBoost

```python
import xgboost as xgb
```

```python
# create instance of XGBoost
xgb_model = xgb.XGBRegressor()
```

```python
# fitting the data to the model
xgb_model.fit(X_train,Y_train)
```

```python
# mean squared error calculation
X_train_prediction = xgb_model.predict(X_train)
training_data_mse = mean_squared_error(Y_train, X_train_prediction)
print("Training data MSE is :", training_data_mse)

X_test_prediction = xgb_model.predict(X_test)
testing_data_mse = mean_squared_error( Y_test, X_test_prediction)
print("Testing data MSE is : ", testing_data_mse)
```

```python
# r2 score calculation
X_train_prediction = xgb_model.predict(X_train)
training_data_r2 = r2_score(Y_train, X_train_prediction)
print("Training data r2_score is :", training_data_r2)

X_test_prediction = xgb_model.predict(X_test)
testing_data_r2 = r2_score( Y_test, X_test_prediction)
print("Testing data r2_score is : ", testing_data_r2)
```
